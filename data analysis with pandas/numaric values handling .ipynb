```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
df.dropna()
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
df.iloc[:,1:3].values
```

```python
imputer.fit(df.iloc[:,1:3].values)
x=imputer.transform(df.iloc[:,1:3].values)
```
