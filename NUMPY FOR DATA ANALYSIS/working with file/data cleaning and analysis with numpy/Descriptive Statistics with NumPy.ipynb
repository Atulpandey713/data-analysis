```python
import numpy as np
```

```python
arr = np.random.randint(low=1, high=101, size=100)
arr

mean =np.mean(arr)
median = np.median(arr)

variance = np.var(arr)
std_deviation = np.std(arr)

q1 = np.percentile(arr,25)
q2 = np.percentile(arr,50)
q3 = np.percentile(arr,75)


iqr = q3 -q1
skewness= np.mean((arr-mean)**3)/(std_deviation**3)
kurtosis= np.mean((arr-mean)**4)/(std_deviation**4)

print('mean :', mean)
print('median :', median)

print("-"*10)

print('variance :',variance)
print('std_deviation :', std_deviation)

print("-"*10)

print('Q1 :', q1)
print('Q2 :', q2)
print('Q3 :', q3)

print('IQR :', iqr)

print("-"*10)
print('skewness :', skewness)
print('kurtosis :', kurtosis)
```
