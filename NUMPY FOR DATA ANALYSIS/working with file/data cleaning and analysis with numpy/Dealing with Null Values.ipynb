```python
import numpy as np
```

## 1 none to np.non

```python
arr = np.array([1,2,3,None,4,5,None])
print(arr)


arr[arr == None] = np.nan
print(arr)

#array([1,2,3,None,4,5,None], dtype=object)
```

```python
arr[arr!=None]
```

## 2 checking the null value

```python
arr = np.array([1,2,3,np.nan,4,5,np.nan])
print(arr)

print(np.isnan(arr))
```

## 3 Removing the null value

```python
arr = np.array([1,2,3,np.nan,4,5,np.nan])
print(np.isnan(arr))
print(arr[np.isnan(arr)])
print(~np.isnan(arr))
print(arr[~np.isnan(arr)])
```

## 4 filling null with fixed value

```python
arr = np.array([1,2,3,np.nan,4,5,np.nan])
print(arr)

arr[np.isnan(arr)]=0
print(arr)
```

## 5 filling the null with average

```python
arr = np.array([1,2,3,np.nan,4,5,np.nan])
print(arr)

arr[np.isnan(arr)]=np.nanmean(arr)
print(arr)
```
