```python
import numpy as np
arr = np.random.randint(1,11,size=(5,5))
arr
```

## 1) Filtering Data:

## 1.1) Filtering Elements Greater Than 5

```python
arr[arr>5]
```

##1.2) Filtering Even ElementS

```python
arr[arr%2 == 0]
```

1.3)Filtering Elements Meeting Multiple Conditions: 

```python
arr[(arr>5) & (arr%2 == 0)]
```

1.4) Filtering Elements Based on Division BY EITHER 3 OR 7

```python
arr[(arr % 3 == 0) | (arr % 7 == 0)]
```

1.5) Filtering Based on Conditions of Another Array 

```python
arr1 = np.array([True, False, True,False,True])
arr[arr1]
```

1.6) Filtering Based on Specific Row Conditions:

```python
filtered_arr = arr[2][arr[2,:] > 5]
filtered_arr
```

## 2) Aggregating Data:

2.1) summation

```python
arr = np.array([1,2,3,4,5])
print(np.sum(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.sum(arr))
print(np.sum(arr, axis = 0))
print(np.sum(arr, axis = 1))
```

2.2) mean

```python
arr = np.array([1,2,3,4,5])
print(np.mean(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.mean(arr))
print(np.mean(arr, axis = 0))
print(np.mean(arr, axis = 1))
```

2.3) standerd deviation

```python
arr = np.array([1,2,3,4,5])
print(np.std(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.std(arr))
print(np.std(arr, axis = 0))
print(np.std(arr, axis = 1))
```

2.4) Minimum and Maximum:

```python
print("minimum")
arr = np.array([1,2,3,4,5])
print(np.min(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.min(arr))
print(np.min(arr, axis = 0))
print(np.min(arr, axis = 1))

print('-'*20)

print("maximum")
arr = np.array([1,2,3,4,5])
print(np.max(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.max(arr))
print(np.max(arr, axis = 0))
print(np.max(arr, axis = 1))
```
