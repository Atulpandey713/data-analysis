```python
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

## CATEGORY WITH COUNT

```python
for name in df['Category'].unique():
  ct = 0
  for i in df['Category']:
    if(i == name):
      ct += 1
  print(name,":", ct)
```

## CATEGORY WITH COUNT USING DECTIONARY

```python
Categories = {}
for name in df['Category'].unique():
  ct = 0
  for i in df['Category']:
    if(i == name):
      ct += 1
  Categories[name] = ct
```

```python
Categories
```

```python
for i in df['Category'].unique():
  print(i,Categories[i])
```
