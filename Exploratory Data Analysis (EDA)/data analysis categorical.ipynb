Q 2:- TOTAL APP IN ART_AND_DESIGN

```python
c = 0
for i in df['Category']:
  if (i == 'ART_AND_DESIGN'):
    c += 1
print("ttotal",c,"application are ART_AND_DESIGN category")
```

Q 4:- TOTA NUMBER OF FREE AND PAID APPLICATION 

```python
#FOR FREE APPLICATION
f = 0
for i in df['Type']:
  if (i == 'Free'):
    f += 1
print("total",f," free application are ART_AND_DESIGN category",end=" ")

## for paid application
p = 0
for i in df['Type']:
  if (i == 'Paid'):
    p += 1
print(" and total",p," paid application are ART_AND_DESIGN category")
```

Q 4:- TOTAL PERCENTAGE OF FREE APPLICATION

```python
print("total",int((f/(p+f))*100),"% free application")
```
