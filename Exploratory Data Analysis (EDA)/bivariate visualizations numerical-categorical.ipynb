```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

BAR POLT OF DATA

```python
sns.barplot(x="class",y="age",data=df)
```

```python
sns.barplot(x="class",y="age",data=df, estimator="sum")
```

```python
sns.barplot(x="class",y="fare",data=df, estimator="sum")
```

```python
sns.barplot(x="class",y="fare",hue="sex",data=df)
```

```python
sns.barplot(x="class",y="fare", hue="sex",data=df, estimator="sum")
```

BOX PLOT OF DATA

```python
sns.boxplot(x="class",y="age",data=df)
```

```python
sns.boxplot(x="class",y="age",hue="survived",data=df)
```

VIOLIN PLOT OF DATA

```python
sns.violinplot(x="class",y="age",data=df)
```

DISTPLOT OF DATA

```python
sns.distplot(df[df["class"]=="second"]["age"])
```
